```python
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.stats
import numpy as np
import seaborn as sns
```

```python
plt.rcParams['figure.figsize'] = (5,4)
plt.rcParams['axes.titlesize'] = 'xx-large'
plt.rcParams.update({
    k: 'xx-large' for k in plt.rcParams 
    if (
        k.endswith('labelsize') or
        k.endswith('fontsize')
    )
})
```

```python
f_N = scipy.stats.norm.pdf
```

```python
colors = sns.color_palette()
```

```python
fig = plt.figure()
x = np.linspace(0, .5, 1000)
lw=3
with plt.xkcd():
    plt.plot(x, f_N(x - .1, 0, .02), color=colors[0], label='$\sum p(t-t^A_k)$', lw=lw)
    plt.plot(x, f_N(x - .3, 0, .03), color=colors[3], label='$\sum p(t-t^C_k)_{Z=0}$', lw=lw)
    plt.plot(x, f_N(x - .3, 0, .035)*.8, color=colors[3], ls='--', label='$\sum p(t-t^C_k)_{Z=1}$', lw=lw)
    plt.axvspan(0,.05, alpha=.15, color='cyan')
    sns.despine()
    plt.xticks(
        [0, .05, .145, .235, .37, .5], 
        ['$0$', '$\\tau^A_0$', '$\\tau^A_1$', '$\\tau^C_0$', '$\\tau^C_1$', '$\Delta t$'])
    plt.yticks([])
    plt.ylabel('Probability density')
plt.legend(frameon=False, bbox_to_anchor=(0.8,1), borderaxespad=0)
fig.savefig('../manuscript/sketch-detailed.svg')
```

```python
fig = plt.figure()
x = np.linspace(0, .5, 1000)
lw=3
with plt.xkcd():
    plt.plot(x, f_N(x - .1, 0, .02), color=colors[0], label='A', lw=lw)
    plt.plot(x, f_N(x - .1, 0, .023), color=colors[2], label='B', lw=lw)
    plt.plot(x, f_N(x - .3, 0, .03), color=colors[3], label='C', lw=lw)
    plt.axvspan(0,.05, alpha=.15, color='cyan')
    sns.despine()
    plt.xticks(
        [1], 
        ['time'])
    plt.yticks([])
    plt.ylabel('Probability density')
    plt.legend(frameon=False, bbox_to_anchor=(0.8,1), borderaxespad=0)
fig.savefig('../manuscript/sketch-simple.svg')
```
