```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f = lambda x: 1 / (1 + np.exp(-(x - 5)))
```

```python
N = 10000
```

```python
t = np.arange(0,N)
```

```python
plt.plot(t,f(t))
```

```python
d = np.random.normal(.3,1,N)
```

```python
x = [0]
X = [0]
treshold = 5
for i in range(N):
    prod = []
    for j in range(i):
        prod.append(f(j)**X[i - j])
        if X[i - j] == 1:
            break
    prod = np.prod(prod)
    x_temp = (x[i] + d[i])
    x_next = x_temp * prod
    if x_next >= treshold:
        X.append(1)
        x.append(0)
    else:
        X.append(0)
        x.append(x_next)
x, X = np.array(x), np.array(X)
```

```python
plt.plot(x)
plt.plot(X*treshold, marker='|', ls='none')
plt.xlim(0,100)
```
